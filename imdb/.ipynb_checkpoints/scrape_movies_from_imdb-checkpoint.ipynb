```python
from bs4 import BeautifulSoup as bs
import requests, openpyxl
```

```python
#config language when extracting the info
headers = {'Accept-Language': 'en-US,en;q=0.8'}

#create excel file
excel = openpyxl.Workbook()
#show sheet names
print(excel.sheetnames)
#set sheet active and change the sheet name
sheet = excel.active
sheet.title = "top250_movies"
print(excel.sheetnames)

#write info to sheet
sheet.append(['Movie Rank', 'Movie Name', 'Year of Released', 'IMDB Rating'])



try:
    #access the web link
    source = requests.get("https://www.imdb.com/chart/top/?ref_=nv_mv_250", headers = headers)
    #show link's state (the link is accessed successfully or show the error)
    source.raise_for_status()
    
    #extract html source code of page
    soup = bs(source.text, 'html.parser')
    
    #find the html tag contains all info about movies
    #parent -> child nodes
    movies = soup.find('tbody', class_ = "lister-list").find_all('tr')
    
    #extract each movie info
    for movie in movies:
        name = movie.find('td', class_ = "titleColumn").a.text
        rating = movie.find('td', class_ = "ratingColumn imdbRating").strong.text
        year = movie.find('td', class_ = "titleColumn").span.text.strip('()')
        rank = movie.find('td', class_ = "titleColumn").get_text(strip=True).split('.')[0]

        sheet.append([rank, name, year, rating])

except Exception as e:
    print(e)

    
excel.save("E:\\STUDY\\ANALYTICS PROJECTS\\PRACTICE\\Scrape Portfolio\\top250movies.xlsx")
```
